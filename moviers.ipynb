Feature Selection Part

```python
movies.columns
```

```python
movie_details =movies
```

```python
movie_details
```

```python
movies=movies[['genres', 'id', 'keywords', 'overview', 'tagline', 'title', 'cast', 'director' ]]
```

```python
movies
```

```python
movies['tags'] = movies['genres']+movies['keywords']+movies['overview']+movies['tagline']+movies['cast']+movies['director']
```

```python
movies
```

```python
new_data  = movies.drop(columns=['overview', 'genres', 'keywords', 'tagline', 'cast', 'director'])
```

```python
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=4803, stop_words='english')
```

```python
cv
```

```python
vector= cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['title']=="The Godfather"].index[0]
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda x:x[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
def recommend(movies):
    index=new_data[new_data['title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x:x[1])
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].title)
```

```python
recommend("Spider-Man")
```
